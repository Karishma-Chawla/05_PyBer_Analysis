```python
#Set the x-axis &tics
x_axis=np.arange(len(pyber_ride_df))
tick_locations=[value for value in x_axis]

# Calculate stdev
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
stdev

# Plot the graph with tick locations
pyber_ride_df.plot.bar(x="Month",y="Avg. Fare ($USD)",color="skyblue",yerr=stdev,capsize=3,label = "Avg. Fare($USD)")
plt.xticks(x_axis,pyber_ride_df["Month"])
plt.ylim(0,60,5)
plt.show()
```
